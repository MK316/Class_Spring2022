# Text Preprocessing:

```python
# word_tokenize: separating punctuation as individual word items
words = word_tokenize(text)
print('Total words including punctuation: ', len(words))
```

```python
# Without punctuation:
wordlist = []

for w in words:
  if len(w) > 1:
    wordlist.append(w)
print('Total words: ', len(wordlist))
print(wordlist)
```

# To lower case & remove stop words

```python
# Lower case
lowerword = []

for w in wordlist:
  w1 = w.lower()
  lowerword.append(w1)
print(lowerword)
```

```python
from nltk.corpus import stopwords  
nltk.download('stopwords')
print(words)
```

```python
# Removing stopwords: 'the' 'a' 'to' 'from' etc.

words = [w for w in words if not w in stopwords.words('english')]
print(words)
print(len(words))
```

# Word frequency

```python
fd = nltk.FreqDist(words)
fd
print('Word type: %d'%len(fd),'개')
```

```python
fwords = nltk.FreqDist(words).most_common(422)
fwords[:5]
```

```python
# One time occurrences:
freq_num1 = [item[1] for item in fwords]
num1 = freq_num1.count(1); 
print('Number of words that occur only once: %d'%num1)
print('The rest: %d'%(422-num1))
```

```python
# nth word that occurs more than 2
i=0
for n in freq_num1:
  if n >2:
#    print(n)
    i += 1
print('몇 번째 단어까지 2회 초과 빈도? %d'%i, '번째')
```

```python
# Words that occur more than twice
fwords = nltk.FreqDist(words).most_common(25)
fwords
```

```python
# If you know 24 words, approximately 1/4 (24.17%) is covered.
psum = sum(freq_num1[:24])
print(psum)
print("Percentage of top 24 words: %d"%((psum/422)*100),'%')
```

```python
# Low frequency words

fd2 = nltk.FreqDist(words).most_common()[-20:]
fd2
```
