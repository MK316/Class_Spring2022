```python
#@markdown ðŸ”„ Making functions: etts('text'), ktts('text'), btts('text)
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "us",
                    slow = False)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def btts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "co.uk",
                    slow = False)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ctts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = "ca",
                    slow = False)
  
  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")
```

# ðŸŒ± Introduction

```python
Intro = "2" #@param = [1,2,3,4]
n = int(Intro) -1
txt = ["Hello students! I'll be giving you a brief introduction to a special vowel in English. The schwa vowel is a central-mid-unrounded vowel sound that occurs in many languages. In English, the schwa vowel is the most common vowel sound and often appears in unstressed syllables. It is sometimes referred to as a 'neutral' or 'reduced' vowel sound because it lacks distinctive-features when compared to other vowel sounds.",
"The connection between the schwa vowel and English stress lies in the fact that unstressed-syllables in English are usually pronounced with the schwa sound. Stress in English is the prominence or emphasis given to certain syllables within a word or phrase. It is essential to the correct pronunciation and understanding of words in English, as stress patterns can alter the meaning of words or distinguish between different grammatical forms. For example, I'll record the record's performance.",
"In multisyllabic words, some syllables are stressed, which means they are more prominent. On the other hand, others are unstressed and they are less prominent. Unstressed syllables are often pronounced with a schwa vowel because they are shorter in duration, quieter, and lower in pitch than stressed syllables. This reduced pronunciation helps to maintain the natural rhythm and balance of stress patterns in spoken English.",
"In summary, the schwa vowel is connected to English stress because it typically appears in unstressed syllables, providing a reduced, neutral vowel sound that contrasts with the more prominent stressed syllables."]
mytxt = txt[n]
etts(mytxt)
Audio("E-audio.mp3", autoplay = True)
```

#A. Examples in different word positions

```python
positions = "1_word-initial" #@param = ["1_word-initial", "2_word-medial", "3_word-final"]
pos = positions.split("_")
num = pos[0]
n=int(num)-1

wordlist = ["ago, away, along, amaze, contain, suppose, asleep, upon, arrive, obtain",
            "agony, relative, holiday, buffalo, telephone, company, belief, jacket, cement, open, typical, cousin, capital, support, believe, signal, famous, nation, certain",
            "soda, sofa, zebra, papa"]
wd = wordlist[n]
intro2 = pos[1]
mytxt = "Let me say some English words appearing in %s. Listen how the schwa vowel is pronounced. "%intro2 + wd

etts(mytxt)
Audio("E-audio.mp3", autoplay = True)
```

# C. Stressed vs. unstressed vowel duration 

```python
#@markdown "mechanic, tomorrow, banana"
language = "ko" #@param = ["en","ko"]

txt = "Listen to the following words: mechanic, tomorrow, banana"

if language == "en":
  etts(txt)
elif language == "ko":
  ktts(txt)

Audio("E-audio.mp3", autoplay=True)
```
